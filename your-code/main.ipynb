# Probability

## Challenge 1
There is a box with 10 white balls, 12 red balls and 8 black balls. Calculate the probability of:
* Taking a white ball out.
* Taking a white ball on first extraction and a black ball out on second extraction without replacing the first one.
* Taking a white ball on first extraction and a black ball out on second extraction replacing the first one.
* Taking 3 balls and them being all different colours.

#### Challenge 1 — Probability

##### Given:
- White balls = 10  
- Red balls = 12  
- Black balls = 8  

$$
\text{Total balls} = 10 + 12 + 8 = 30
$$

----

#### 1. Probability of taking a white ball

$$
P(\text{White}) = \frac{10}{30} = \frac{1}{3}
$$

----

#### 2. White first, then black (without replacement)

- First draw (white):
$$
P(W_1) = \frac{10}{30}
$$

- Second draw (black | first was white):
$$
P(B_2 \mid W_1) = \frac{8}{29}
$$

- Combined probability:
$$
P(W_1 \cap B_2) = \frac{10}{30} \cdot \frac{8}{29} = \frac{8}{87}
$$

----

#### 3. White first, then black (with replacement)

- First draw:
$$
P(W_1) = \frac{10}{30}
$$

- Replace → total remains 30

- Second draw:
$$
P(B_2) = \frac{8}{30}
$$

- Combined probability:
$$
P(W_1 \cap B_2) = \frac{10}{30} \cdot \frac{8}{30} = \frac{4}{45}
$$

----

#### 4. Three balls, all different colours

We want:
- 1 White
- 1 Red
- 1 Black

##### Step 1: Favorable outcomes

$$
\binom{10}{1} \cdot \binom{12}{1} \cdot \binom{8}{1} = 960
$$

##### Step 2: Total outcomes

$$
\binom{30}{3} = 4060
$$

##### Step 3: Probability

$$
P(\text{all different}) = \frac{960}{4060} = \frac{48}{203}
$$

---

#### Final Answers

- $P(\text{White}) = \frac{1}{3}$
- $P(\text{White then Black, no replacement}) = \frac{8}{87}$
- $P(\text{White then Black, with replacement}) = \frac{4}{45}$
- $P(\text{all different colours}) = \frac{48}{203}$

## Challenge 2

You want to go on picnic on Saturday. You hate rain, and would know to plan your weekend. To help you make a decision, you gather the following data about rainy days:

* 40% of all days are cloudy.
* About 50% of cloudy days are rainy. 
* About 20% of non-cloudy days are rainy.

What is the chance of rain during next Saturday?

#### Challenge 2 — Probability of Rain

##### Given:

- $P(\text{Cloudy}) = 0.4$
- $P(\text{Rain} \mid \text{Cloudy}) = 0.5$
- $P(\text{Rain} \mid \text{Not Cloudy}) = 0.2$

$$
P(\text{Not Cloudy}) = 1 - 0.4 = 0.6
$$

---

##### Step 1: Law of Total Probability

$$
P(\text{Rain}) =
P(\text{Rain} \mid \text{Cloudy}) \cdot P(\text{Cloudy})
+
P(\text{Rain} \mid \text{Not Cloudy}) \cdot P(\text{Not Cloudy})
$$

---

##### Step 2: Substitute values

$$
P(\text{Rain}) =
(0.5 \cdot 0.4) + (0.2 \cdot 0.6)
$$

$$
= 0.2 + 0.12 = 0.32
$$

---

##### Final Answer

$$
P(\text{Rain on Saturday}) = 0.32
$$

## Challenge 3

A and B are events of a probability such that $P(A) = 0.3$, $P(B) = 0.6$ and $P(A \cap B) = 0.1$

Which of the following statements are true?  
**HINT: draw a big Venn Diagram by your side**  
* $P(\overline{A}) = 0.7$
* $P(A \cup B) = 0.6$
* $P(A \cap \overline{B}) = 0.2$
* $P(A \cup \overline{A}) = 1$  
* $P(A \cap (B \cup \overline{B})) = 0.4$
* $P(\overline{A} \cap \overline{B}) = 0.3$
* $P(\overline{A \cap B}) = 0.9$

#### Challenge 3 — Probability Statements

##### Given:

- $P(A) = 0.3$
- $P(B) = 0.6$
- $P(A \cap B) = 0.1$

----

##### Step 1: Useful derived values

##### Complement of A
$$
P(\overline{A}) = 1 - P(A) = 1 - 0.3 = 0.7
$$

##### Union of A and B
$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$
$$
= 0.3 + 0.6 - 0.1 = 0.8
$$

##### A and not B
$$
P(A \cap \overline{B}) = P(A) - P(A \cap B)
$$
$$
= 0.3 - 0.1 = 0.2
$$

##### Complement of both (neither A nor B)
$$
P(\overline{A} \cap \overline{B}) = 1 - P(A \cup B)
$$
$$
= 1 - 0.8 = 0.2
$$

##### Complement of intersection
$$
P(\overline{A \cap B}) = 1 - P(A \cap B)
$$
$$
= 1 - 0.1 = 0.9
$$

----

##### Step 2: Evaluate statements

##### 1. $P(\overline{A}) = 0.7$ TRUE

----

##### 2. $P(A \cup B) = 0.6$  FALSE  
Actual value:
$$
P(A \cup B) = 0.8
$$

----

##### 3. $P(A \cap \overline{B}) = 0.2$  TRUE

----

##### 4. $P(A \cup \overline{A}) = 1$  TRUE  
(A or not A = entire sample space)

----

##### 5. $P(A \cap (B \cup \overline{B})) = 0.4$  FALSE  

Since:
$$
B \cup \overline{B} = S
$$

Then:
$$
P(A \cap S) = P(A) = 0.3
$$

----

##### 6. $P(\overline{A} \cap \overline{B}) = 0.3$  FALSE  
Actual value:
$$
P(\overline{A} \cap \overline{B})=0.2
$$

----

##### 7. $P(\overline{A \cap B}) = 0.9$  TRUE

----

##### Final Answers Summary

| Statement | Result |
|----------|--------|
| $P(\overline{A}) = 0.7$ |  True |
| $P(A \cup B) = 0.6$ |  False |
| $P(A \cap \overline{B}) = 0.2$ |  True |
| $P(A \cup \overline{A}) = 1$ |  True |
| $P(A \cap (B \cup \overline{B})) = 0.4$ |  False |
| $P(\overline{A} \cap \overline{B}) = 0.3$ |  False |

## Challenge 4

 * If someone flies to China, the probability of getting Covid is 1%  
 * If someone flies to Italy, the probability of getting Covid is 20%  

I roll 1 dice with numbers 1, 2, 3, 4, 5, 6:  
 * If 1: I fly to Italy
 * If not 1: I fly to China

During my trip, I got Covid. What is the probability that I flew to China?

#### Challenge 4 — Bayes' Theorem

##### Given:

- $P(\text{Italy}) = \frac{1}{6}$
- $P(\text{China}) = \frac{5}{6}$

- $P(\text{Covid} \mid \text{Italy}) = 0.2$
- $P(\text{Covid} \mid \text{China}) = 0.01$

----

##### Step 1: Total probability of getting Covid

$$
P(\text{Covid}) =
P(\text{Covid} \mid \text{Italy}) \cdot P(\text{Italy})
+
P(\text{Covid} \mid \text{China}) \cdot P(\text{China})
$$

$$
= (0.2 \cdot \tfrac{1}{6}) + (0.01 \cdot \tfrac{5}{6})
$$

$$
= \tfrac{0.2}{6} + \tfrac{0.05}{6}
= \tfrac{0.25}{6}
$$

----

##### Step 2: Apply Bayes’ Theorem

We want:

$$
P(\text{China} \mid \text{Covid})
=
\frac{
P(\text{Covid} \mid \text{China}) \cdot P(\text{China})
}{
P(\text{Covid})
}
$$

----

##### Step 3: Substitute values

$$
P(\text{China} \mid \text{Covid})
=
\frac{
0.01 \cdot \tfrac{5}{6}
}{
\tfrac{0.25}{6}
}
$$

$$
=
\frac{0.05/6}{0.25/6}
=
\frac{0.05}{0.25}
= 0.2
$$

----

##### Final Answer

$$
P(\text{China} \mid \text{Covid}) = 0.2
$$

## BONUS

You roll 2 dice. What is the probability that the sum of the numbers is 7? And 8? And 9? Then simulate 10000 times and draw a countplot.

your explanation here

##### BONUS — Dice Probabilities & Simulation

##### Problem

Roll 2 fair dice. Find:

- $P(\text{sum} = 7)$  
- $P(\text{sum} = 8)$  
- $P(\text{sum} = 9)$  

Then simulate 10,000 rolls and visualize the results.

----

##### Step 1: Sample Space

Each die has 6 outcomes:

$$
\text{Total outcomes} = 6 \times 6 = 36
$$

All outcomes are equally likely.

----

##### Step 2: Count favorable outcomes

##### Sum = 7

Possible pairs:

- (1,6), (2,5), (3,4), (4,3), (5,2), (6,1)

$$
\text{Count} = 6
\quad \Rightarrow \quad
P(7) = \frac{6}{36} = \frac{1}{6}
$$

----

##### Sum = 8

- (2,6), (3,5), (4,4), (5,3), (6,2)

$$
\text{Count} = 5
\quad \Rightarrow \quad
P(8) = \frac{5}{36}
$$

----

##### Sum = 9

- (3,6), (4,5), (5,4), (6,3)

$$
\text{Count} = 4
\quad \Rightarrow \quad
P(9) = \frac{4}{36} = \frac{1}{9}
$$

----

##### Final Probabilities

- $P(7) = \frac{1}{6} \approx 0.167$
- $P(8) = \frac{5}{36} \approx 0.139$
- $P(9) = \frac{1}{9} \approx 0.111$

----

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# simulate 10,000 rolls
np.random.seed(42)
dice1 = np.random.randint(1, 7, 10000)
dice2 = np.random.randint(1, 7, 10000)

sums = dice1 + dice2

# create dataframe
df = pd.DataFrame({'sum': sums})

# plot
plt.figure()
sns.countplot(x='sum', data=df)

plt.title("Distribution of Dice Sums (10,000 simulations)")
plt.xlabel("Sum")
plt.ylabel("Count")

plt.show()

# show proportions
print(df['sum'].value_counts(normalize=True).sort_index())
```
